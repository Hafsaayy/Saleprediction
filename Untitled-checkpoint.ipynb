```python
def Date_Format(Dset,column):
    Dset=Dset.copy()
    Dset[column] = pd.to_datetime(Dset[column])
    Dset[column + '_Year'] = Dset[column].apply(lambda x :x.year)
    Dset[column + '_Months'] = Dset[column].apply(lambda x :x.month)
    Dset[column + '_Days'] = Dset[column].apply(lambda x :x.day)
    Dset=Dset.drop(column,axis=1)
    
    return Dset
def onehot_encode(Dset,column):
    Dset=Dset.copy()
    dummies=pd.get_dummies(Dset[column],prefix=column)
    Dset=pd.concat([Dset,dummies],axis=1)
    Dset=Dset.drop(column, axis=1)
    return Dset
```

```python
Dset=DS
Dset=Dset.copy()
#Droping unnecessary Column
Dset=Dset.drop(['Row ID','Country','Product Name','Customer Name'],axis=1)

#Droping Customer Specific Feature Column
Dset=Dset.drop(['Order ID','Customer ID'],axis=1)

#Date Feature
Dset=Date_Format(Dset,column='Order Date')
Dset=Date_Format(Dset,column='Ship Date')



#onehot encoding categorial features
for column in ['Ship Mode', 'Segment','City','State','Postal Code','Region','Product ID','Category','Sub-Category']:
    Dset=onehot_encode(Dset,column=column)

```
